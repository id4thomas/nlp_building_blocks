```python
import json
```

```python
def process_log(log_name):
    with open(f"results/{log_name}.json", "r") as f:
        data = json.load(f)


    result = data["result"]
    model_name = result["model_id"]
    n = result["num_prompts"]
    concurrency = result["max_concurrency"]
    
    prompt = 4096
    completion = 1024
    print(f"\tModel {model_name}, Concurrency {concurrency}")
    print(f"\t{n} Requests (Prompt {prompt} Completion {completion})")
    
    rps = result["request_throughput"]
    
    # Time Calculation
    ## per 1K (in seconds)
    time_per_1k_request = int(1000/rps)
    print(f"\tRPS {rps:.3f}")
    print(f"\tTime per 1K Req: {time_per_1k_request}s ({time_per_1k_request/60:.2f}M, {time_per_1k_request/3600:.2f}H)")
    
    # TPS
    output_tps = result["output_throughput"]
    total_tps = result["total_token_throughput"]
    print(f"\tTPS Output {int(output_tps)} Total {int(total_tps)}")
    
```

```python
log_names = [
    "qwen3_5_35b_a3b_bf16-c1", # Qwen3.5-35B-A3B C8
    "qwen3_5_35b_a3b_bf16-c4", # Qwen3.5-35B-A3B C8
    "qwen3_5_35b_a3b_bf16-c8", # Qwen3.5-35B-A3B C8
]
```

```python
for log_name in log_names:
    print(log_name)
    process_log(log_name)
```
