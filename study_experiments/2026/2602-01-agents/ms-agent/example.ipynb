```python
from builder import AgentBuilder
```

```python
agent = AgentBuilder.build(
    name="WeatherAgent",
    user_id="1234",
    tool_names=["location", "weather"],
    user_name="id4thomas"
)
```

```python
query="내 현재 위치와 주변 도시들을 알려주고 그 도시들의 날씨들도 알려줘."
# query="서울 주변 도시들을 알려주고 그 도시들의 날씨들도 알려줘. 도시 명칭은 반환값 그대로 사용해"
# query="도구를 사용해서 서울 날씨 알려줘. 도구의 결과도 json 그대로 알려줘"

result = await agent.run(
    query,
    session=None,
    options={"store": False}
)
print(result)
```

```python
for i, message in enumerate(result.messages):
    contents = message.contents
    print(f"MSG {i} {message.role} - {len(contents)} contents")
    for cont_i, content in enumerate(contents):
        print(f"CONT {cont_i} - {content.type}")
        if content.type=='text':
            print(f"\t{repr(str(content))}")
        elif content.type=='function_result':
            print(f"\t{content.name}\t{content.result}")
        else:
            print(f"\t{content.name}\t{content.arguments}")
    print('-'*30)
```
