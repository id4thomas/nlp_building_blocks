# modu-2020

```python
def process_entity(e):
    label = e['label'][:2]
    
    return {
        'begin': e['begin'],
        'end': e['end'],
        'label': label,
        'detail': e['label']
    }
    
```

```python
data20_dir = os.path.join(source_dir, "2020/20_150tags_NamedEntity/EXNE2102209160")
fnames = [x for x in os.listdir(data20_dir) if "json" in x]
print(len(fnames))
```

```python
data = {
    'id': [], # documentId
    'source': [], # source (for tracking)
    'text': [], # sentences List[str]
    'entities': []
}

for fname in tqdm(fnames):
    # 'utf-8-sig': Unexpected UTF-8 BOM (decode using utf-8-sig): line 1 column 1 (char 0)
    # BOM: Byte Order Mark (BOM)
    with open(os.path.join(data20_dir, fname), "r", encoding='utf-8-sig') as f:
        sample = json.load(f)
        
    file_id = sample['id']
    
    for doc in sample['document']:
        document_id = doc['id']
        sentences = [x['form'] for x in doc['sentence']]
        entities = [
            [process_entity(e) for e in x['NE']]
            for x in doc['sentence']
        ]
        
        data['id'].append(document_id)
        data['source'].append(f'modu-20/{file_id}')
        data['text'].append(sentences)
        data['entities'].append(entities)

data_df = pd.DataFrame.from_dict(data)
```
