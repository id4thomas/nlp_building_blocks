```python
import json
import os
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
root_dir = os.path.dirname(os.path.dirname(os.getcwd()))
print(root_dir)
```

```python
data_dir = os.path.join(root_dir, f"data/source/modu")
print(os.listdir(data_dir))
```

# modu-2020

```python
data20_dir = os.path.join(data_dir, "2020/20_150tags_NamedEntity/EXNE2102209160")
fnames = [x for x in os.listdir(data20_dir) if "json" in x]
print(len(fnames))
```

```python
## Sentence Lengths
num_sentences = []
sentence_lengths = []
document_lengths = []

for fname in tqdm(fnames):
    # 'utf-8-sig': Unexpected UTF-8 BOM (decode using utf-8-sig): line 1 column 1 (char 0)
    # BOM: Byte Order Mark (BOM)
    with open(os.path.join(data20_dir, fname), "r", encoding='utf-8-sig') as f:
        sample = json.load(f)
    
    
    for doc in sample['document']:
        doc_sent_lens = [len(x['form']) for x in doc['sentence']]
        num_sentences.append(len(doc_sent_lens))
        sentence_lengths.extend(doc_sent_lens)
        document_lengths.append(sum(doc_sent_lens))
```

```python
print(sentence_lengths[:50])
print(document_lengths[:50])
```

```python
print(f"Document Num Sents avg {sum(num_sentences)/len(num_sentences):.2f} max {max(num_sentences)}")
print(f"Sentence Count {len(sentence_lengths)} avg {sum(sentence_lengths)/len(sentence_lengths):.2f} max {max(sentence_lengths):.2f}")
print(f"Document Count {len(document_lengths)} avg {sum(document_lengths)/len(document_lengths):.2f} max {max(document_lengths):.2f}")
```

```python
plt.clf()
plt.xlim(0,500)
plt.hist(sentence_lengths, bins=30)
plt.title("[MODU-20] Sentence Length Dist")
plt.show()
```

```python
plt.clf()
plt.xlim(0, 1000)
plt.hist(document_lengths, bins=50)
plt.title("[MODU-20] Document Length Dist")
plt.show()
```
